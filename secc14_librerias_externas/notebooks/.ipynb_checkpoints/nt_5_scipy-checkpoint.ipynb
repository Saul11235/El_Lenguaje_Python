# Explorando algunas caracteristicas de Matplotlib

 <img src="https://matplotlib.org/stable/_static/logo_light.svg" alt="Logo de matplotlib" width="400">



```python
# importando matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Gráfico de líneas
print("1. Gráfico de líneas")
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Seno de x', color='b', linestyle='-', linewidth=2)
plt.title('Gráfico de Líneas', fontsize=14)
plt.xlabel('Eje X', fontsize=12)
plt.ylabel('Eje Y', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Gráfico de barras
print("\n2. Gráfico de barras")
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 33]

plt.figure(figsize=(8, 5))
plt.bar(categorias, valores, color=['#FF6347', '#98C9E8', '#FFD700', '#32CD32', '#FF1493'])
plt.title('Gráfico de Barras', fontsize=14)
plt.xlabel('Categorías', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.show()
```

```python
# 3. Gráfico de dispersión
print("\n3. Gráfico de dispersión")
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='purple', alpha=0.6, edgecolors='w', s=100)
plt.title('Gráfico de Dispersión', fontsize=14)
plt.xlabel('Eje X', fontsize=12)
plt.ylabel('Eje Y', fontsize=12)
plt.show()
```

```python
# 4. Histograma
print("\n4. Histograma")
data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title('Histograma', fontsize=14)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 5. Gráfico de subgráficos (Subplots)
print("\n5. Gráfico de Subgráficos")
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(8, 10))
axs[0].plot(x, y1, label='Seno de x', color='r')
axs[0].set_title('Gráfico de Seno')
axs[0].legend()

axs[1].plot(x, y2, label='Coseno de x', color='g')
axs[1].set_title('Gráfico de Coseno')
axs[1].legend()

plt.tight_layout()  # Para evitar superposición de los subgráficos
plt.show()
```
