# ¿Que son los Notebooks?

Estos son entornos para ejecutar y documentar código python, estos entornos se dividen en celdas
las celdas son::

|  Tipo de celda   |                Descripcion                 |
|------------------|--------------------------------------------|
| Code             | Código (python en este caso a ejecutarse   |
| Markdown         | Celdas anotativas                          |
| Raw              | Texto plano                                |


```python
for x in range(10):
    print(f"saludo desde codigo python en linea n° {x+1}")
```

# En un notebook pueden haber muchas celdas
En un notebook, pueden haber muchas celdas de varios tipos
algunos modulos externos pueden estar optimizados para trabajar con notebooks, incluso widgets que ofrecen diversas opciones.
https://ipywidgets.readthedocs.io/en/stable/ 

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 33]

# Crear el gráfico
fig, ax = plt.subplots(figsize=(8, 6))

# Crear el gráfico de barras
bars = ax.bar(categorias, valores, color=['#FF6347', '#98C9E8', '#FFD700', '#32CD32', '#FF1493'])

# Añadir etiquetas y título
ax.set_xlabel('Categorías', fontsize=14, color='black')
ax.set_ylabel('Valores', fontsize=14, color='black')
ax.set_title('Gráfico de Barras Bonito', fontsize=16, color='#4B0082')

# Añadir las etiquetas encima de cada barra
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 2, round(yval, 2), ha='center', va='bottom', fontsize=12, color='black')

# Establecer estilo de la cuadrícula
ax.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
